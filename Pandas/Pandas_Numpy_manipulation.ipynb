```python
## Renaming columns
df = df.rename(columns={'ship_date':'Ship_ Date'})
df.head()
```

```python
df['New Quantity'] = df['quantity'].apply(lambda x: x*2)
df.head()
```

```python
## Group by clause
grouped_agg = df.groupby('region')['sales'].agg(['mean','sum','count'])
grouped_agg
```

```python
df1 = pd.DataFrame({'Key': ['A','B','C'], 'Value1': [1,2,3]})
df2 = pd.DataFrame({'Key': ['A','B','C'], 'Value2': [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merging and joining
pd.merge(df1,df2, on="Key", how="inner")
```

```python
pd.merge(df1,df2, on="Key", how="outer")
```
